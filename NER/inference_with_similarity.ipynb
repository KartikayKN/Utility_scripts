```python
# def match_key_value_pairs(key_list, value_list, bracket_array):
#     key_value_dict = {'key_values': {}, 'misc': {}, 'bracket': {}}
#     for key, value, bracket in zip(key_list, value_list, bracket_array):
#         if value == "":
#             continue
#         dbKey, _ = match_similarity(key)
#         if dbKey is not None and dbKey != 'misc':
#             key_value_dict['key_values'][dbKey] = value
#             key_value_dict['bracket'][dbKey] = bracket
#         else:
#             key_value_dict['misc'][key] = value
#             key_value_dict['bracket'][key] = bracket

#     return key_value_dict


def match_similarity(word, threshold=0.75):
    matched_key = None
    matched_prob = 0
    word = re.sub(':', '', word)
    for key, key_options in key_dict.items():
        max_prob = 0
        for value in key_options:
            prob = difflib.SequenceMatcher(None, word.lower(), value.lower()).ratio()
            if prob > threshold and max_prob < prob:
                max_prob = prob
        if matched_prob < max_prob:
            matched_prob = max_prob
            matched_key = key

    return matched_key, matched_prob

def get_standard_key(key):
    matched_key, matched_prob = match_similarity(key)
    if matched_key is not None and matched_key != 'misc':
        return True , matched_key
    return False , key if matched_key is None else matched_key

def standardize_keys(key_val_map):
    standard_dic = {}
    misc_dic = {}
    for key in list(key_val_map):
        for x in list(key_val_map[key]):
            if_standard_key , standard_key =  get_standard_key(x)
            if if_standard_key:
                if key not in standard_dic.keys():
                    standard_dic[key] = {}
                standard_dic[key][standard_key] = key_val_map[key][x]
            else:
                if key not in misc_dic.keys():
                    misc_dic[key] = {}
#                 standard_dic[key][standard_key] = key_val_map[key][x]
                misc_dic[key][standard_key] = key_val_map[key][x]        
    return standard_dic ,misc_dic



duplicates_label_list = ['fax', 'email', 'ATTN', 'contact' , 'phoneNo']

def get_key_val_map(standard_dic):
    dic = {}
    for key in standard_dic.keys():
        for x in standard_dic[key].keys():
            if x not in dic.keys() and x not in duplicates_label_list:
                dic[x] = standard_dic[key][x]
    return dic
```
