```python
import xmltodict
import json
import glob
from Json2PascalVoc.Converter import Converter
myConverter = Converter()
PATH = "C:/Users/KartikayGupta/Xml To Yolo/L/clean_xml/"


def preprocess(dic):
    arr = []
    if(isinstance(dic['annotation']['object'],dict)):
        dic['annotation']['object'] = [dic['annotation']['object']]
    for obj in dic['annotation']['object']:
        if obj['name'] == "Keyvalue" or obj['name']=="key value" or obj['name']== "Copy of keyvalue" or obj['name'] == "keyvalue":
            obj['name'] = "keyvalue"
            arr.append(obj)
        if obj['name'] == "heaing" or obj['name'] == "heading":
            obj['name'] = "heading"
            arr.append(obj)        
    dic['annotation']['object'] = arr
    dic['annotation']['path'] = PATH + dic['annotation']['filename'] 
    return dic

dic = {'data' : []}


for doc in glob.glob("C:/Users/KartikayGupta/Xml To Yolo/L/xm/*.xml"):
    with open(doc) as file:
        file_data = file.read()
        dict_data = xmltodict.parse(file_data)
        print(doc)
        dict_data= preprocess(dict_data)
        dic['data'].append(dict_data)

with open("sample.json", "w") as outfile:
    json.dump(dic, outfile)
    
myConverter.convertJsonToPascal("sample.json")
```
